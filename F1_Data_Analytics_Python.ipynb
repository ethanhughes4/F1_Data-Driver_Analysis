```python
drivers.columns = drivers.columns.str.lower()
constructors.columns = constructors.columns.str.lower()
races.columns = races.columns.str.lower()
results.columns = results.columns.str.lower()
qualifying.columns = qualifying.columns.str.lower()
circuits.columns = circuits.columns.str.lower()
```

```python
from sqlalchemy import create_engine
username = "postgres"
password = "Tox!cteddy1"
host = "localhost"
port = "5432"
database = "F1_Data_Analysis"
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# Load tables into PostgreSQL
drivers.to_sql(
    "drivers",
    engine,
    if_exists="replace",
    index=False
)

constructors.to_sql(
    "constructors",
    engine,
    if_exists="replace",
    index=False
)

races.to_sql(
    "races",
    engine,
    if_exists="replace",
    index=False
)

results.to_sql(
    "results",
    engine,
    if_exists="replace",
    index=False
)

qualifying.to_sql(
    "qualifying",
    engine,
    if_exists="replace",
    index=False
)

circuits.to_sql(
    "circuits",
    engine,
    if_exists="replace",
    index=False
)

print("All tables loaded successfully.")
```
